```python
import pandas as pd
```

```python
original = pd.read_csv('https://s3.amazonaws.com/far-right/fourchan/chan_example.csv', parse_dates=['created_at'])
```

```python
df = original
```

```python
df.head()
```

```python
text = df['com'].str.replace('\<wbr\>','')
```

```python
extraction = {
    'link': '(https?://.*?(?=( |(\<br\>)|(\<\/span\>))))'
    , 'quoted_post': 'href="#p([0-9]*)"'
}
```

```python
for name, regex in extraction.items():
    
    temp = text.str.extractall(regex).reset_index()
    temp = temp[['level_0',0]]
    temp.columns = ['row','extraction']
    
    df[name]=temp.groupby('row')['extraction'].apply(list)
```

```python
find_replace = {
     '<a href="(\/[A-Za-z0-9]{0,4}\/thread/[0-9]*)(#p)[0-9]*" class="quotelink">?(&gt;)+[0-9]+(</a>)*(<br>)*(t\.)*': ''
    , '<a href="(#p)?[0-9]*" class="quotelink">?(&gt;)+[0-9]+(</a>)*(<br>)*(t\.)*': ''
    , '<a href="(.*)" class="quotelink">?(&gt;)+/()\/1/<\/a>':''
    , '&quot;': '\"'
    , '\&#039': '\''
    , '<br>': ' '
    , '<span class="quote">(&gt;)+(t\.)?': '~~'
    , '</span>': '~~'
    , '&lt;': '<'
    , '&gt;': '>'
    ,'\';' : "'"
    }
```

```python
for keys, values in find_replace.items():
    text = text.str.replace(keys, values)
```

```python
df['text'] = text
```

```python
clean = df.to_dict(orient='records')
```
